# Chapter 1 Abstract Data Types
## 1.1 Introduction 
* **Simple data types: consist of values that are in the most basic forms and cannot be decompoesed into smaller parts, i.e., integers and real rtpes.**
* **Complex data types: are constructed of multiple components consisting of simple types or other comples types**

### 1.1.1 Abstraction
* **Abstraction: a mechanism for separating the properties of an object and restricting the focus to those relevant in the current context**
    * **Procedural abstraction: the use of a function or method knowing what it does but ignoring how it's accomplished**
    * **Data abstraction: the separation of the properties of a datatype (its values and operations) from the implementationb of that data type.**


### 1.1.2 Abstract Data Types
* **Abstract data type(ADT) is a programmer-defined data type that specifies a set of data values and a collection of well-defined operations that can be performed on those values**
    * **The separation is typically enforced by requiring interaction witg the abstract data type though an *interface* or defined set of operations. This is known as *information hiding*.**
    * ***constructors*: create an initialize new instances of the ADT.**
    * ***accessors*: return data contained in an instance without modifying it**
    * ***Mutators*: modify the contents of an ADT instance**
    * ***Iterators*: process individual data components sequentally**

### 1.1.3 Data Structures
* ***Simple* ADT: is composed of a single or several individually named data fields such as those used to represent a date or rational number**
* ***Complex ADTs* are composed of a collection of data values such as the Python list or dictionary**
    * **Complex abstract data types are implemented using a particular *data structure*, which is the physical representation of how data is organized and manipulated.**
    * **Data structures can be characterized by how they store and organize the individual data elements and what operations are availiable for accessing and manipulating the data**

### 1.1.4 General Definitions
* **Collection: a group of values with no implied organization or relationship between the individual values**
* **A *container* is any data structure or abstrac data types that stores and organizes a collection. The individual values of he collection are known as *elements* of the container and a continaer with no elemnets is said to be *empty*.**
* **A *sequence* is a container in which the elements are arranged in linear order from front to back, with each element accessible by position**
* **A *sorted sequence* is one in which the position of the elements is based on a presribed relationship between each element and its successor**
* **In computer science, the term list is commonly used to refer to any collection with a linear ordering. The ordering is such that every element in the collection, except the first one, has a unique predecessor and every element, except the last one, has a unique successor.**

## 1.3 Bags
* **A *bag* is a simple container that can be used to store a collection of items.**
    * **The bag container restrict access to the individual items by only definining operations for adding and removing inidvidual items, for determining if an item is in the bag, and for traversing over the collection of items.**

### 1.3.1 The Bag Abstract Data Type
* **A bag is a container that stores a collection in which duplicate values are allowed. The items, each of which is individually stored, have no particular order but they must be comparable**
* bag(): **creates a bag that is initially empty**
* length(): **returns the number of items stored in the bag. Accessed using the** len() **function**
* containers(item): **determines if the given target item is stored in the bag and returns the appropriate boolean value. Accessed using the** in **operator**
* add(item): **adds the given item to the bag**
* remove(item): **removes and returns an occurence of** item **from the bag. An exception is raised if the element is not in the bag**
* iterator(): **creates and returns an iterator that can be used to iterate over the collection of items.**

#### Why a Bag ADT?
1. **Focus on solving the probelm at hand instead of worrying about the implementation of the container**
1. **Reduce the chance of introducing errors from misuse of the list since it provides additional operations that are not appropriate for a bag**
1. **Provide better coordination between different modules and designers**
1. **Easily swap out our current implementation of the Bag ADT for a different, posibly more efficient version later**

### 1.3.2 Selecting a Data Structure
1. **Does the data structure procide for the storage requirments as specified by the domain of the ADT**
1. **Does the data structure provide the necessary data access and manipulation funcionality to implement the ADT**
1. **Does the data structure lend itseld to efficient implementation of the operation**
    * **List: can store any type of comparabel object, including duplicates. Each item is stored individually, including duplicates, which means the reference to eahc individual ovehct is stored and later accessible when needed**
        * **A list stores references to objects.
    * **Dictionary: stores the key/value paris in which the key component must be comparable and unique.**

### 1.4.1 Designing an Iterator
* **To use Python's traversal mechansim with bag ADT, we must define an iterator class, which is a class in Python containing two special methods**, \_\_iter\_\_ **and** \_\_next\_\_.
* **Iterator classes are commonly defined in the same module as the corresponding class**
* **Constructure defines two data fiels:**
    * **An alias to the list used to store the items in the bag**
    * **The other is a loop index variable that will be used to iterate over the list(initalized to 0)**
* **The** \_\_iter\_\_ **method returns a reference to the object itself and is always implmented to do so**
* **The** \_\_next\_\_ **is caleed to return the next item in the container. The method saves a reference to the current item indicated by the loop variable. The loop is then incremented by one to prepare it for the next invocation of the** \_\_next\_\_ **method. If there are no additional itesm, the method raise a** StopIteration **exception that flags the** for **loop to terminate**
* **We must also add an** \_\_iter\_\_ **method to our** Bag **class.**
    * **Theis method, which is responsible for creating and returning an instance of the** \_Bagiterator **class, is automatically called at the beginning of the** for **loop to create an iterator object for use with the loop construct**

```
class _BagIterator:
    def __init__( self, theList ):
        self._bagItems = theList
        self._curItem = 0
        
    def __iter__( self ):
            return self
        
    def __next__( self ):
        if self._curItem < len( self._bagItems ):
            item = self._bagItems[ self._curItem ]
            self._curItem += 1
            return item
        else:
            raise StopIteration
```
